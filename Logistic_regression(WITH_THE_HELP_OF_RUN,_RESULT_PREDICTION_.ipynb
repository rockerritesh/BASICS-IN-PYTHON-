```
def gradient_descent(y,x):
    m_curr = 0.000001    #you can change this for better prediction
    n_curr= 10     #you can change this for better prediction
    ileteration=1000   #you can change this for better prediction
    n=len(x)
    learning_rate=0.3    #you can change this for better prediction
    
    for i in range(ileteration):
        y_predicted=1/(1+np.exp(-(n_curr+x*m_curr)))  #hypothetical function
        cost = -(1/n)*sum([val for val in (-y*np.log1p(y_predicted)-((1-y)*np.log1p(1-y_predicted)))])  #cost function
        md = -(1/n)*sum(x*(y-y_predicted))
        nd = -(1/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate * nd
        n_curr = n_curr - learning_rate * md
        print ("m {}, n {} ,cost {}, iteration {} ".format(m_curr,n_curr,cost, i))
        
        
    return m_curr,n_curr 
```

```
m,c=gradient_descent(y,x)
```
