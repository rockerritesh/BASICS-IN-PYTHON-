```

y = np.array([300000,500000,700000]) #enter the price of house #you can pass matrix or group of matrix
x = np.array([1000,2000,3000])  #enter the area of house
b = np.array([1,2,3])  #eneter the room of each house
```

```
#our equation for this problem is 
#  y = c +m*x +n*b
#here we know value of y,x,b this might be the price of house , land area, no of room
#
```

```
def gradient_descent(y,x,b):
    m_curr =242 
    n_curr =1
    c_curr =2  #initial supossed value of m,n&c
    ileration = 100000 #number of time the function will run
    n = len(x)
    learning_rate = 0.000000003  #most important term which decide how the prediction will start by takiing large or small  baby steps
    
    for i in range(ileration):
        y_predicted = c_curr  + m_curr*x + n_curr*b   #hypotical function or actual linear function we consider
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])  #cost function j(0o,01)
        md = -(1/n)*sum(x*(y-y_predicted)) #derivative of cost function with respect to 01
        cd = -(1/n)*sum(y-y_predicted)  #derivative of cost function with respect  to 0
        nd = -(1/n)*sum(b*(y-y_predicted))#derivative of cost function with respect  to 02
        m_curr = m_curr - learning_rate * md  #predicting value of m and updating in each illetration
        n_curr = n_curr - learning_rate * nd  #predicting value of n and updating value in each iletration
        c_curr = c_curr - learning_rate* cd  #predicting value of c and updating value in each iletration
        print ("m {}, n {}, c {} ,cost {}, iteration {} ".format(m_curr,n_curr,c_curr,cost, i)) #you can use any of them this will show you steps
    #print(f'm ={m_curr},n = {n_curr},c = {c_curr},cost = {cost},iteration = {i}') #this will not show you steps 

    return m_curr,n_curr,c_curr

  
```

```
m_curr,n_curr,c_curr = gradient_descent(y,x,b)
```

```
xa = int(input('enter the area of land of which you want to predict'))
ba  =int( input('enter the no of rooom of that house which you want to predict'))
price =  c_curr  + m_curr*xa + n_curr*ba
print(f'pridected price is {price} ')
```
