## newton rapson method

```python
def function(x):
  return x*x*x*x+3*x*x-50
```

```python
def funcder(x):
  return 4*x*x*x + 6*x
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=np.linspace(-10,10,100)
plt.plot(w,function(w))
plt.plot(w,funcder(w))
plt.grid(True)
```

```python
x=-10 #point where the roots have positive value
xi=x
x0=[]
```

```python
while abs(function(x)) > 0.001:
  x1=x-(function(x)/funcder(x))
  x0.append(x1)
  x=x1
  xf=x #xf is the final roots of function
x
```

## least square

### linear line

```python
num=input("enter the number of data set you have:")
num=int(num)
num
```

```python
num
xset=np.linspace(1,10,10)
yset=np.linspace(3,12,10)
noise=np.random.rand(10)+np.random.rand(10)
yset=yset+noise
yset
```

```python
'''for i in range(num):
  x=float(input("enter x value"))
  xset.append(x)
  y=float(input("enter y value"))
  yset.append(y)'''  #use only you want to use manual input method insted of we have linespace as ploting linear line
```

```python
sumx=0
sumy=0
sumxy=0
sumxx=0
for i in range(num):
  sumx+=xset[i]
  sumy+=yset[i]
  sumxx+=xset[i]*xset[i]
  sumxy+=xset[i]*yset[i]
```

```python
b=((sumy*sumxx)-(sumx*sumxy))/((num*sumxx)-(sumx*sumx))
a=((num*sumxy)-(sumx*sumy))/((num*sumxx)-(sumx*sumx))
print(a)
print(b)
```

```python
X=np.array([[sumxx,sumx],[sumx,num]])
Y=np.array([[sumxy],[sumy]])
[a,b]=np.linalg.inv(X)@Y
print(a)
print(b)
```

```python
g=np.linspace(1,10,10)
h=[]
for i in range(len(g)):
  d=a*g[i]+b
  h.append(d)
h  
```

```python
plt.rcParams["figure.figsize"]=(20, 13)
plt.plot(g,h,'bo--')  #produced data
plt.plot(xset,yset,'go') #input data
```

```python
sumx=0
sumy=0
sumxy=0
sumxx=0
for i in range(num):
  '''sumx+=np.log(xset[i])
  sumy+=np.log(yset[i])
  sumxx+=np.log(xset[i])*np.log(xset[i])
  sumxy+=np.log(xset[i])*np.log(yset[i])'''
  sumx+=xset[i]
  sumy+=np.log(yset[i])
  sumxx+=xset[i]*xset[i]
  sumxy+=xset[i]*np.log(yset[i])
```

```python
b=((sumy*sumxx)-(sumx*sumxy))/((num*sumxx)-(sumx*sumx))
a=((num*sumxy)-(sumx*sumy))/((num*sumxx)-(sumx*sumx))
a=np.exp(a)
print(a)
print(b)
```

```python
X=np.array([[sumxx,sumx],[sumx,num]])
Y=np.array([[sumxy],[sumy]])
[a,b]=np.linalg.inv(X)@Y
a=np.exp(a)
print(a)
print(b)
```

```python
g=np.linspace(1,10,10)
h=[]
for i in range(len(g)):
  d=a*np.exp(g[i]*b)
  h.append(d)
h  
```
